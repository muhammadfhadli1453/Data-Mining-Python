## 1. Import package

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

## 2. Read data

```python
df = pd.read_csv("Example2.csv")
```

```python
df
```

## 3. Declare and input new data

```python
new_data = {'Umur':'Muda',
            'Pendapatan':'Menengah',
            'Pelajar':'Ya',
            'Credit Rating':'Fair'}
```

```python
df = df.append(new_data, ignore_index=True)
```

```python
X = df[df.columns[:-1]]
y = df['Membeli Komputer']
```

```python
X
```

```python
y
```

## 4. Encode string in data

```python
le = LabelEncoder()
X = X[X.columns[:]].apply(le.fit_transform)
```

```python
X
```

## 5. Training and predict

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X[:-1], y[:-1])
```

```python
result = clf.predict([X.iloc[-1]])
```

```python
print("Class dari ", new_data, " adalah ", result[0])
```
