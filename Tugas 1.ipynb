## 1. Import package

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

## 2. Create data

```python
X = np.array([[1, 2], 
              [1, 4], 
              [1, 0],
              [10, 2], 
              [10, 4], 
              [10, 0]])
```

## 3. Plot data

```python
plt.title('Data Awal')
plt.scatter(X[:,0], X[:,1])
plt.show()
```

## 4. Clustering

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

## 5. Plot clustering result

```python
plt.title('KMeans Clustering')
plt.scatter(X[:,0], X[:,1], c=list(kmeans.labels_))
plt.show()
```

## 6. Create and plot new data

```python
X_test = np.array([[0, 0], [12, 3]])
```

```python
plt.title('Data Baru')
plt.scatter(X[:,0], X[:,1], c=list(kmeans.labels_))
plt.scatter(X_test[:,0], X_test[:,1])
plt.show()
```

## 7. Predict new data and plot

```python
cluster = kmeans.predict(X_test)
```

```python
plt.title('Jawaban Tugas 1')
plt.scatter(X[:,0], X[:,1], c=list(kmeans.labels_))
plt.scatter(X_test[:,0], X_test[:,1], c=list(cluster))
plt.show()
```
